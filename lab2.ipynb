```python
## Q1 Need to load 2 seperate matrics, x=observations, y=output(prices) ,rank of the feuture matrix, and pseudo inverse to find the cost of each product
import pandas as pd
from numpy.linalg import matrix_rank,pinv
import numpy as np
file_path='Lab2 Session Data.xlsx'
df=pd.read_excel(file_path,sheet_name=0)
A=df[['Candies (#)','Mangoes (Kg)','Milk Packets (#)','Payment (Rs)']].to_numpy()
print("The rank of matrix A(feature+output):",matrix_rank(A))
X=df[['Candies (#)','Mangoes (Kg)','Milk Packets (#)']].to_numpy()
print("The observation matrix is(Feature Columns):\n",X)
print("The dimentionility of X:",X.shape)
print("The number of vectors in the matrix X:",X.shape[1])
print("The rank of matrix X(feature):",matrix_rank(X))
Y=df[['Payment (Rs)']].to_numpy()
print("The result/output matrix is(Output):\n",Y)
print("The dimentionility of Y:",Y.shape)
print("The number of vectors in the matrix Y:",Y.shape[1])
print("The cost of each product:",pinv(Y))


# Q2 All the customer with price above 200-Rich and others as poor
from sklearn.tree import DecisionTreeClassifier
x=df[['Payment (Rs)']]
y=np.where(df['Payment (Rs)']>200,'Rich','Poor')
clf = DecisionTreeClassifier()
clf.fit(x,y)
print("Predict all the customers who are rich/poor:",clf.predict(x))
```
