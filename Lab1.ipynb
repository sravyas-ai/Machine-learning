```python
# q1

s=input("Enter a strings: ") # an input string with vowels and consonents
vowels="aeiouAEIOU"
countv=0 # to count the number of vowels
countc=0 # to count the number of consonents

for ch in s:
    if ch in vowels:
        countv+=1
    else:
        countc+=1

print("Vowels:",countv)
print("Consonants:",countc)
```

```python
#q2

import numpy as np # using it to make the matric calculation easier

r1 = int(input("Enter rows of matrix A: ")) 
c1 = int(input("Enter columns of matrix A: "))
A = np.array([list(map(int, input().split())) for _ in range(r1)]) # putting both the row and the coloumn numbers in an array

r2 = int(input("Enter rows of matrix B: "))
c2 = int(input("Enter columns of matrix B: "))
B = np.array([list(map(int, input().split())) for _ in range(r2)])

if c1!=r2:
    print("Error") # checks if they are multiplicable , if they are the same they are multiplicable
else:
    product = np.dot(A, B) # multiplying A and B
    print("Product AB =\n", product)
```

```python
#q3

l1=input("Enter first list: ")
l2=input("Enter second list: ")
list1=l1.split() #we are splitting an input string
list2=l2.split()

common = set(list1)&set(list2) # adding only the common elements
print("Common elements:", common)
print("Number of common elements:", len(common))
```

```python
# q5

import random
from statistics import mean, median, mode # a matematical function import funtion for mean mode and medium in statistics
nums = [random.randint(100, 150) for _ in range(100)] #takes random numbers from the range of 100-150
print("Numbers:", nums)
print("Mean:", mean(nums))
print("Median:", median(nums))
print("Mode:", mode(nums))
```
