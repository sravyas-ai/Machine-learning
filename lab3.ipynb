```python
# q1) dot product,length of vectors in their ecludium form
import numpy as np
a=np.array([1,4,2,7,4])
b=np.array([5,8,3,9,3])

def dot_prod(a,b): 
    n=len(a)
    product=sum(a[i]*b[i] for i in range(n)) # mulitplies the numbers of each row and sums it up to give the dot product
    return product

def length(a,b):
    length1=np.sqrt(sum(a[i]**2 for i in range(len(a)))) #Its squares the ech number in the vector and sums it up to get the length of euclidium norm
    length2=np.sqrt(sum(b[i]**2 for i in range(len(b))))
    return length1,length2
l1,l2=length(a,b)
print("Results for dot product: ",dot_prod(a,b),np.dot(a,b))
print("Results for the length in euclidium norm:\nFor vector A: ",l1,np.linalg.norm(a),"\nFor vector B:",l2,np.linalg.norm(b))
```
